```python
import streamlit as st
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

st.title("TTC Delay Analytics Dashboard")

df = pd.read_csv("../../data/processed/ttc_feature_engineered_final.csv")

# Sidebar filters
line_filter = st.sidebar.multiselect("Line", df["line"].unique())
station_filter = st.sidebar.multiselect("Station", df["station"].unique())

filtered_df = df.copy()

if line_filter:
    filtered_df = filtered_df[filtered_df["line"].isin(line_filter)]

if station_filter:
    filtered_df = filtered_df[filtered_df["station"].isin(station_filter)]

# KPIs
col1, col2, col3 = st.columns(3)

col1.metric("Total Delays", len(filtered_df))
col2.metric("Average Delay", round(filtered_df["min_delay"].mean(),1))
col3.metric("High Delay %",
            round((filtered_df["min_delay"]>=20).mean()*100,1))

# Top delay causes
st.subheader("Top Delay Causes")

cause_counts = filtered_df["code_reduced"].value_counts().head(10)

fig, ax = plt.subplots()
sns.barplot(x=cause_counts.values, y=cause_counts.index, ax=ax)

st.pyplot(fig)

# Delay by hour
st.subheader("Delays by Hour")

hour_counts = filtered_df.groupby("hour").size()

fig, ax = plt.subplots()
hour_counts.plot(ax=ax)

st.pyplot(fig)
```
